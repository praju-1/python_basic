# 1. Write a Python Program to Check if a Number is Positive, Negative or Zero?

```python
try:
    n = int(input("Enter any number : "))
    def check_num_type(num : int):
        "This function will return number is Positive, Negative or Zero"
        try:
            if num > 0:
                print("Number is positive")
            elif num < 0:
                print("Number is negative")
            else:
                print("Number is Zero")
        except Exception as e:
            print(e)
    check_num_type(n)
except Exception as e:
        print(e)
```

# 2. Write a Python Program to Check if a Number is Odd or Even?

```python
try:
    n = int(input("Enter any number : "))
    def odd_even(num : int):
        try:
            if num % 2 == 0:
                print("The given number is even")
            
            else:
                print("The given number is odd")
            
        except Exception as e:
            print(e)
        
    odd_even(n)
        
except Exception as e:
        print(e)
```

# 3. Write a Python Program to Check Leap Year?

```python
try:
    year = int(input("Enter the year : "))
    def check_leap_year(year : int):
        "This function check whether leap year"
        try :
            print("\nThe year is : ", year)
            if (((year % 4 == 0) and (year % 100 != 0)) or (year % 400 == 0)):
                return True
            else:
                return False
        except Exception as e:
            print(e)

    leap_year = check_leap_year(year)
    if(leap_year):
        print("\nGiven year is Leap year. ")
    else:
        print("\nGiven year is not Leap year.")
except Exception as e:
            print(e)
```

# 4. Write a Python Program to Check Prime Number?

```python
n = int(input("Enter a number: "))

try:
    def prime_number(num : int):
        try:
            flag = False
            if num == 1:
                print(num, "is not a prime number")
            elif num > 1:
                # check for factors
                for i in range(2, num):
                    if (num % i) == 0:
                        # if factor is found, set flag to True
                        flag = True
                        # break out of loop
                        break

                # check if flag is True
                if flag:
                    print(num, "is not a prime number")
                else:
                    print(num, "is a prime number")
                    
        except Exception as e:
            print(e)
            
    prime_number(n)
    
except Exception as e:
        print(e)
```

# 5. Write a Python Program to Print all Prime Numbers in an Interval of 1-10000?

```python
try:
    #It will take input from user
    a = int(input("Enter the number from you want prime number : "))
    b = int(input("Enter the number upto you want prime number : "))
            
    def prime_number(num1 : int, num2 : int):
        " This function return the total prime number between given range"
        try:
            count = 0
            for element in range(num1, num2):
                #This check condition for prime number
                if element % 2 == 0:
                    print ("The Prime number is  : ", element)
                    count += 1
            print("\nThe total prime numbers between", num1, "and", num2, "is :", count)

        except Exception as e:
            print(e)

    prime_number(a, b)
except Exception as e:
            print(e)
```
